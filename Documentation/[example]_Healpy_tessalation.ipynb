# Modules : 

```python
import healpy as hp 
import numpy as np 
import matplotlib.cm as cm
```

# Initialisation : 

```python
nside_tess = 4 #nside taken to tessalate the healpy map into big sections.
nside_map = 2048 #nseide of the map we are going to cut into bigger patches.
```

# Code : 

## Create an empty map : 

```python
nfields = hp.nside2npix(nside_tess) 
npix = hp.nside2npix(nside_map)
It = int(npix/nfields) # Number of pixels per field, i.e number of pixels in each of the big patches. 

Fmap = np.empty(npix) # create a empty healpy map. 

hp.mollview(map=Fmap, coord=None, unit='', xsize=2000, nest=True, cbar=True, cmap=cm.jet, norm='hist',
            return_projected_map=True)
```

## Select the field : 

```python
Fmap[0:0+It]+= 1 

hp.mollview(map=Fmap, coord=None, unit='', xsize=2000, nest=True, cbar=True, cmap=cm.jet, norm='hist',
            return_projected_map=True)
```

## Select the first six fields : 

```python
i=0
for j in range(0,6*It,It):  
                    
                Fmap[j:j+It]+= i
                i = i+1
                
hp.mollview(map=Fmap, coord=None, unit='', xsize=2000, nest=True, cbar=True, cmap=cm.jet, norm='hist',
            return_projected_map=True)    
```

## Select all the fields itteratively : 

```python
i=0
for j in range(0,len(Fmap),It):  
                    
                Fmap[j:j+It]+= i
                i = i+1
                
hp.mollview(map=Fmap, coord=None, unit='', xsize=2000, nest=True, cbar=True, cmap=cm.jet, norm='hist',
            return_projected_map=True)            
```

## Conclusion : 

Using this technique we have separated a Healpy map of $N_{SIDE}=2048$ into $N_{SIDE}=4$ => $192$ fields each of them containing :  

```python
nfields = hp.nside2npix(nside_tess) 
npix = hp.nside2npix(nside_map)
It = int(npix/nfields) 
print(It)
```
